```python
#This provides the schema of the incoming csv file

from pyspark.sql.types import *
import json

# 日期,時段,進站,出站,人次
# date, hour, origin, destination, passenger_count

jschemacsv = '{"fields":[{"metadata":{},"name":"日期","nullable":true,"type":"string"},{"metadata":{},"name":"時段","nullable":true,"type":"string"},{"metadata":{},"name":"進站","nullable":true,"type":"string"},{"metadata":{},"name":"出站","nullable":true,"type":"string"},{"metadata":{},"name":"人次","nullable":true,"type":"integer"}],"type":"struct"}'

schemacsv = StructType.fromJson(json.loads(jschemacsv))
```

```python
#This provides the schema of the Delta/Parquet table being loaded

from pyspark.sql.types import *
import json

jschematable = '{"fields":[{"metadata":{},"name":"日期","nullable":true,"type":"date"},{"metadata":{},"name":"時段","nullable":true,"type":"string"},{"metadata":{},"name":"進站","nullable":true,"type":"string"},{"metadata":{},"name":"出站","nullable":true,"type":"string"},{"metadata":{},"name":"人次","nullable":true,"type":"integer"},{"metadata":{},"name":"filename","nullable":true,"type":"string"}],"type":"struct"}'

schematable = StructType.fromJson(json.loads(jschematable))
```

```python
#This reads the csv, adds a filename column, and formats the date column, then writes to the Delta/Parquet table

from pyspark.sql.types import *
from  pyspark.sql.functions import *

def loadFullDataFromSource(file_name):
    df = spark.read.format("csv").option("header","true").schema(schemacsv).load("Files/mrtfiles/raw/" + file_name)
    df = df.withColumn("filename", input_file_name())
    df = df.withColumn("日期", to_date(col("日期"), "yyyy-MM-dd")) 
    df.write.mode("append").format("delta").option("mergeSchema", "true").option("schema", "schematable").save("Tables/mrt201912")
    
full_files = [
    '201912.csv'
    ]

for file in full_files:
    loadFullDataFromSource(file)
```
