```python
from pyspark.sql.types import *
import json

# 日期,時段,進站,出站,人次
# date, hour, origin, destination, passenger_count

jschemacsv = '{"fields":[{"metadata":{},"name":"日期","nullable":true,"type":"string"},{"metadata":{},"name":"時段","nullable":true,"type":"string"},{"metadata":{},"name":"進站","nullable":true,"type":"string"},{"metadata":{},"name":"出站","nullable":true,"type":"string"},{"metadata":{},"name":"人次","nullable":true,"type":"integer"}],"type":"struct"}'

schemacsv = StructType.fromJson(json.loads(jschemacsv))
```

```python
from pyspark.sql.types import *
import json

jschematable = '{"fields":[{"metadata":{},"name":"日期","nullable":true,"type":"string"},{"metadata":{},"name":"時段","nullable":true,"type":"string"},{"metadata":{},"name":"進站","nullable":true,"type":"string"},{"metadata":{},"name":"出站","nullable":true,"type":"string"},{"metadata":{},"name":"人次","nullable":true,"type":"integer"},{"metadata":{},"name":"filename","nullable":true,"type":"string"}],"type":"struct"}'

schematable = StructType.fromJson(json.loads(jschematable))
```

```python
from pyspark.sql.types import *
from  pyspark.sql.functions import *

def loadFullDataFromSource(file_name):
    df = spark.read.format("csv").option("header","true").schema(schemacsv).load("Files/unzipped/" + file_name)
    df = df.withColumn("filename", input_file_name())
    #display(df) 
    df.write.mode("append").format("delta").option("mergeSchema", "true").option("schema", "schematable").save("Tables/mrt2019")
    
full_files = [
    '201901.csv',
    '201902.csv',
    '201903.csv',
    '201904.csv',
    '201905.csv',
    '201906.csv',
    '201907.csv',
    '201908.csv',
    '201909.csv',
    '201910.csv',
    '201911.csv',
    '201912.csv'
    ]

for file in full_files:
    loadFullDataFromSource(file)
```
