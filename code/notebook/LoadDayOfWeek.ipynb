```python
from pyspark.sql.types import *
import json

# date, dayofweek

jschemacsv = '{"fields":[{"metadata":{},"name":"date","nullable":true,"type":"string"},{"metadata":{},"name":"dayofweek","nullable":true,"type":"string"}],"type":"struct"}'

schemacsv = StructType.fromJson(json.loads(jschemacsv))
```

```python
from pyspark.sql.types import *
import json

jschematable = '{"fields":[{"metadata":{},"name":"date","nullable":true,"type":"date"},{"metadata":{},"name":"dayofweek","nullable":true,"type":"string"}],"type":"struct"}'

schematable = StructType.fromJson(json.loads(jschematable))
```

```python
from pyspark.sql.types import *
from pyspark.sql.functions import *

df = spark.read.format("csv").option("header","true").schema(schemacsv).load("Files/dayofweek/dayofweek.csv")
df = df.withColumn("date", to_date(col("date"), "MM/dd/yyyy"))
df.write.mode("overwrite").format("delta").option("mergeSchema", "true").option("schema", "schematable").save("Tables/dayofweek")
```
